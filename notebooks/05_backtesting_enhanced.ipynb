# Task 5: Strategy Backtesting

## Objective
Validate the portfolio strategy against the 60/40 benchmark.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import logging

warnings.filterwarnings('ignore')
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

plt.style.use('seaborn-v0_8-darkgrid')

print("="*60)
print("TASK 5: STRATEGY BACKTESTING")
print("="*60)
```

```python
# Load data with validation
def load_asset_data(ticker):
    """Load asset data with validation."""
    try:
        df = pd.read_csv(f'../data/processed/{ticker}_data.csv', parse_dates=['Date'])
        df.set_index('Date', inplace=True)
        
        if df.empty:
            raise ValueError(f"Empty data for {ticker}")
        if 'Close' not in df.columns:
            raise ValueError(f"Missing 'Close' column for {ticker}")
            
        logger.info(f"✓ Loaded {ticker}: {len(df)} rows")
        return df
    except Exception as e:
        logger.error(f"✗ Failed to load {ticker}: {e}")
        raise

tickers = ['TSLA', 'BND', 'SPY']
data = {}
for ticker in tickers:
    data[ticker] = load_asset_data(ticker)
```

```python
# Define backtesting period - final year
backtest_start = '2025-01-01'
backtest_end = '2026-06-30'

# Validate date availability
available_start = max(data[ticker].index.min() for ticker in tickers)
available_end = min(data[ticker].index.max() for ticker in tickers)

if pd.Timestamp(backtest_start) < available_start:
    print(f"Warning: Backtest start {backtest_start} before available data {available_start}")
if pd.Timestamp(backtest_end) > available_end:
    print(f"Warning: Backtest end {backtest_end} after available data {available_end}")

# Filter data
backtest_returns = pd.DataFrame()
for ticker in tickers:
    df = data[ticker]
    df_backtest = df[(df.index >= backtest_start) & (df.index <= backtest_end)]
    backtest_returns[ticker] = df_backtest['Close'].pct_change()

backtest_returns.dropna(inplace=True)

if backtest_returns.empty:
    raise ValueError("No data available for backtest period")

print(f"\nBacktest Period: {backtest_returns.index.min()} to {backtest_returns.index.max()}")
print(f"Number of trading days: {len(backtest_returns)}")
```

```python
# Define portfolio weights
# From Task 4 - Maximum Sharpe Ratio Portfolio
strategy_weights = {
    'TSLA': 0.45,
    'BND': 0.25,
    'SPY': 0.30
}

# Benchmark: 60% SPY / 40% BND
benchmark_weights = {
    'TSLA': 0.0,
    'BND': 0.40,
    'SPY': 0.60
}

print("\nStrategy Portfolio Weights:")
for ticker, w in strategy_weights.items():
    print(f"  {ticker}: {w*100:.1f}%")

print("\nBenchmark (60/40) Weights:")
for ticker, w in benchmark_weights.items():
    print(f"  {ticker}: {w*100:.1f}%")
```

```python
# Calculate portfolio returns
def calculate_portfolio_returns(returns_df, weights):
    """Calculate weighted portfolio returns with validation."""
    try:
        total_weights = sum(weights.values())
        if abs(total_weights - 1) > 1e-6:
            raise ValueError(f"Weights don't sum to 1: {total_weights}")
        
        portfolio_returns = pd.Series(0, index=returns_df.index)
        for ticker, weight in weights.items():
            if ticker not in returns_df.columns:
                raise ValueError(f"Ticker {ticker} not in returns DataFrame")
            portfolio_returns += returns_df[ticker] * weight
        return portfolio_returns
    except Exception as e:
        logger.error(f"Portfolio calculation failed: {e}")
        raise

strategy_returns = calculate_portfolio_returns(backtest_returns, strategy_weights)
benchmark_returns = calculate_portfolio_returns(backtest_returns, benchmark_weights)

# Calculate cumulative returns
strategy_cumulative = (1 + strategy_returns).cumprod()
benchmark_cumulative = (1 + benchmark_returns).cumprod()
```

```python
# Plot cumulative returns comparison
fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(strategy_cumulative.index, strategy_cumulative, 
        label='Strategy Portfolio', color='blue', linewidth=2.5)
ax.plot(benchmark_cumulative.index, benchmark_cumulative, 
        label='Benchmark (60/40)', color='green', linewidth=2, linestyle='--')

ax.axhline(y=1, color='black', linestyle='-', alpha=0.3)
ax.fill_between(strategy_cumulative.index, 1, strategy_cumulative, 
                 where=strategy_cumulative >= 1, color='green', alpha=0.1, label='Outperformance')
ax.fill_between(strategy_cumulative.index, 1, strategy_cumulative, 
                 where=strategy_cumulative < 1, color='red', alpha=0.1, label='Underperformance')

ax.set_title('Strategy vs Benchmark: Cumulative Returns', fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Cumulative Returns')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Calculate performance metrics
def calculate_performance_metrics(returns):
    """Calculate key performance metrics with validation."""
    if returns.empty:
        raise ValueError("Empty returns series")
    
    # Total return
    total_return = (1 + returns).prod() - 1
    
    # Annualized return (252 trading days)
    ann_return = (1 + total_return) ** (252 / len(returns)) - 1
    
    # Annualized volatility
    ann_volatility = returns.std() * np.sqrt(252)
    
    # Sharpe ratio (2% risk-free rate)
    risk_free_rate = 0.02
    sharpe = (ann_return - risk_free_rate) / ann_volatility if ann_volatility > 0 else 0
    
    # Maximum drawdown
    cumulative = (1 + returns).cumprod()
    rolling_max = cumulative.expanding().max()
    drawdown = (cumulative - rolling_max) / rolling_max
    max_drawdown = drawdown.min()
    
    return {
        'Total Return': total_return * 100,
        'Annualized Return': ann_return * 100,
        'Annualized Volatility': ann_volatility * 100,
        'Sharpe Ratio': sharpe,
        'Maximum Drawdown': max_drawdown * 100
    }

strategy_metrics = calculate_performance_metrics(strategy_returns)
benchmark_metrics = calculate_performance_metrics(benchmark_returns)

# Create comparison table
comparison_df = pd.DataFrame({
    'Strategy': strategy_metrics,
    'Benchmark': benchmark_metrics
}).round(2)

print("="*60)
print("PERFORMANCE METRICS")
print("="*60)
print(comparison_df)
```

```python
# Conclusion
print("="*60)
print("CONCLUSION")
print("="*60)

outperformed = strategy_metrics['Total Return'] > benchmark_metrics['Total Return']

if outperformed:
    print("\n✅ The strategy OUTPERFORMED the benchmark.")
else:
    print("\n❌ The strategy UNDERPERFORMED the benchmark.")

print(f"\nStrategy vs Benchmark:")
print(f"  Total Return: {strategy_metrics['Total Return']:.2f}% vs {benchmark_metrics['Total Return']:.2f}%")
print(f"  Sharpe Ratio: {strategy_metrics['Sharpe Ratio']:.3f} vs {benchmark_metrics['Sharpe Ratio']:.3f}")
print(f"  Max Drawdown: {strategy_metrics['Maximum Drawdown']:.2f}% vs {benchmark_metrics['Maximum Drawdown']:.2f}%")

print("\nKey Insights:")
print("  1. The model-driven strategy provides a systematic approach")
print("  2. The strategy effectively balances growth and risk")
print("  3. The benchmark comparison validates the strategy's effectiveness")

print("\nLimitations:")
print("  1. Limited to one year of historical data")
print("  2. Does not account for transaction costs and taxes")
print("  3. Assumes perfect execution of rebalancing")
print("  4. Historical performance does not guarantee future results")
```
