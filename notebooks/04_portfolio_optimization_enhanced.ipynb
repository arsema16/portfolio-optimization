# Task 4: Portfolio Optimization with Modern Portfolio Theory

## Objective
Construct an optimal portfolio using forecasted TSLA returns and historical BND/SPY returns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy.optimize import minimize
import logging

warnings.filterwarnings('ignore')
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')

print("="*60)
print("TASK 4: PORTFOLIO OPTIMIZATION")
print("="*60)
```

```python
# Load and validate data
def load_data_with_validation(tickers):
    """Load all asset data with validation."""
    data = {}
    for ticker in tickers:
        try:
            df = pd.read_csv(f'../data/processed/{ticker}_data.csv', parse_dates=['Date'])
            df.set_index('Date', inplace=True)
            
            # Validate
            required_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
            missing = [c for c in required_cols if c not in df.columns]
            if missing:
                raise ValueError(f"Missing columns for {ticker}: {missing}")
            if df.empty:
                raise ValueError(f"Empty data for {ticker}")
            
            data[ticker] = df
            logger.info(f"✓ Loaded {ticker}: {len(df)} rows")
        except Exception as e:
            logger.error(f"✗ Failed to load {ticker}: {e}")
            raise
    return data

tickers = ['TSLA', 'BND', 'SPY']
data = load_data_with_validation(tickers)
```

```python
# Calculate returns and validate
returns_df = pd.DataFrame()

for ticker in tickers:
    returns_df[ticker] = data[ticker]['Close'].pct_change()

# Remove NaN and check for empty
returns_df.dropna(inplace=True)
if returns_df.empty:
    raise ValueError("Returns DataFrame is empty after dropping NaN")

print(f"\nReturns data: {len(returns_df)} rows")
print(returns_df.head())
```

```python
# Expected returns: Combine forecasted TSLA with historical BND/SPY
historical_returns = returns_df.mean() * 252

# Forecasted TSLA return (from Task 3)
tsla_forecast_return = 0.15  # ~15% annualized from forecast

expected_returns = pd.Series({
    'TSLA': tsla_forecast_return,
    'BND': historical_returns['BND'],
    'SPY': historical_returns['SPY']
})

# Covariance matrix (annualized)
cov_matrix = returns_df.cov() * 252

print("\nExpected Returns (Annualized):")
print(expected_returns)
print("\nCovariance Matrix:")
print(cov_matrix)
```

```python
# Portfolio performance calculation with validation
def validate_weights(weights):
    """Validate that weights sum to 1 and are non-negative."""
    if len(weights) != 3:
        raise ValueError(f"Expected 3 weights, got {len(weights)}")
    if abs(np.sum(weights) - 1) > 1e-6:
        raise ValueError(f"Weights don't sum to 1: {np.sum(weights)}")
    if np.any(weights < 0):
        raise ValueError(f"Negative weights found: {weights}")
    return True

def portfolio_performance(weights, expected_returns, cov_matrix):
    """Calculate portfolio performance with validation."""
    weights = np.array(weights).flatten()
    try:
        validate_weights(weights)
    except ValueError as e:
        logger.error(f"Weight validation failed: {e}")
        raise
    
    returns = np.sum(expected_returns * weights)
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe = returns / volatility if volatility > 0 else 0
    return returns, volatility, sharpe
```

```python
# Generate efficient frontier
def generate_efficient_frontier(expected_returns, cov_matrix, n_portfolios=10000):
    """Generate random portfolios with validation."""
    n_assets = len(expected_returns)
    results = []
    
    for i in range(n_portfolios):
        try:
            weights = np.random.random(n_assets)
            weights /= np.sum(weights)
            ret, vol, sharpe = portfolio_performance(weights, expected_returns, cov_matrix)
            results.append({
                'Return': ret,
                'Volatility': vol,
                'Sharpe': sharpe,
                'weights': weights
            })
        except Exception as e:
            logger.warning(f"Portfolio {i} failed: {e}")
            continue
    
    if not results:
        raise ValueError("No valid portfolios generated")
    
    return pd.DataFrame(results)

frontier_df = generate_efficient_frontier(expected_returns, cov_matrix, 10000)
print(f"\nGenerated {len(frontier_df)} valid portfolios")
```

```python
# Find optimal portfolios
def find_optimal_portfolios(frontier_df):
    """Find max Sharpe and min volatility portfolios."""
    max_sharpe_idx = frontier_df['Sharpe'].idxmax()
    min_vol_idx = frontier_df['Volatility'].idxmin()
    
    max_sharpe = frontier_df.loc[max_sharpe_idx]
    min_vol = frontier_df.loc[min_vol_idx]
    
    return max_sharpe, min_vol

max_sharpe, min_vol = find_optimal_portfolios(frontier_df)

print("\n" + "="*60)
print("OPTIMAL PORTFOLIOS")
print("="*60)

print("\nMaximum Sharpe Ratio Portfolio:")
print(f"  Return: {max_sharpe['Return']*100:.2f}%")
print(f"  Volatility: {max_sharpe['Volatility']*100:.2f}%")
print(f"  Sharpe Ratio: {max_sharpe['Sharpe']:.4f}")
for i, ticker in enumerate(tickers):
    print(f"  {ticker}: {max_sharpe['weights'][i]*100:.1f}%")

print("\nMinimum Volatility Portfolio:")
print(f"  Return: {min_vol['Return']*100:.2f}%")
print(f"  Volatility: {min_vol['Volatility']*100:.2f}%")
print(f"  Sharpe Ratio: {min_vol['Sharpe']:.4f}")
for i, ticker in enumerate(tickers):
    print(f"  {ticker}: {min_vol['weights'][i]*100:.1f}%")
```

```python
# Plot Efficient Frontier
fig, ax = plt.subplots(figsize=(14, 8))

# Scatter plot
scatter = ax.scatter(frontier_df['Volatility'] * 100, 
                     frontier_df['Return'] * 100,
                     c=frontier_df['Sharpe'], 
                     cmap='viridis', 
                     alpha=0.5, 
                     s=10)

# Mark optimal portfolios
ax.scatter(max_sharpe['Volatility'] * 100, max_sharpe['Return'] * 100,
           color='red', s=300, marker='*', label='Max Sharpe Ratio', edgecolors='black')
ax.scatter(min_vol['Volatility'] * 100, min_vol['Return'] * 100,
           color='blue', s=300, marker='o', label='Min Volatility', edgecolors='black')

# Mark individual assets
for i, ticker in enumerate(tickers):
    ax.scatter(np.sqrt(cov_matrix.iloc[i, i]) * 100, 
               expected_returns[i] * 100,
               s=300, marker='s', label=ticker, edgecolors='black', linewidths=2)

ax.set_xlabel('Volatility (Risk) %', fontsize=12)
ax.set_ylabel('Expected Return %', fontsize=12)
ax.set_title('Efficient Frontier with Optimal Portfolios', fontsize=14, fontweight='bold')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Sharpe Ratio')

plt.tight_layout()
plt.show()
```

```python
# Final Recommended Portfolio
print("="*60)
print("FINAL PORTFOLIO RECOMMENDATION")
print("="*60)

recommended = max_sharpe

print("\nRecommended Portfolio: Maximum Sharpe Ratio")
print("\nWeights:")
for i, ticker in enumerate(tickers):
    print(f"  {ticker}: {recommended['weights'][i]*100:.1f}%")

print(f"\nExpected Annual Return: {recommended['Return']*100:.2f}%")
print(f"Expected Volatility: {recommended['Volatility']*100:.2f}%")
print(f"Sharpe Ratio: {recommended['Sharpe']:.4f}")

print("\nJustification:")
print("  1. This portfolio offers the best risk-adjusted returns")
print("  2. It balances growth (TSLA), stability (BND), and diversification (SPY)")
print("  3. The allocation leverages TSLA's growth potential while managing risk")
```
