# Task 3: Future Forecasting with Confidence Intervals

## Objective
Generate 6-12 month TSLA forecasts with confidence intervals and visual distinction between historical, test predictions, and future forecasts.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
import logging

warnings.filterwarnings('ignore')
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

plt.style.use('seaborn-v0_8-darkgrid')

print("="*60)
print("TASK 3: FUTURE FORECASTING")
print("="*60)
```

```python
# Load and validate data
def load_and_validate_data(filepath):
    """Load data with validation checks."""
    try:
        df = pd.read_csv(filepath, parse_dates=['Date'])
        df.set_index('Date', inplace=True)
        
        # Validate required columns
        required_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
        missing_cols = [col for col in required_cols if col not in df.columns]
        if missing_cols:
            raise ValueError(f"Missing required columns: {missing_cols}")
        
        # Check for empty data
        if df.empty:
            raise ValueError("DataFrame is empty")
        
        logger.info(f"✓ Loaded {len(df)} rows from {filepath}")
        return df
    except Exception as e:
        logger.error(f"Error loading data: {e}")
        raise

try:
    tsla_data = load_and_validate_data('../data/processed/TSLA_data.csv')
    print(f"\nTSLA Data:")
    print(f"  Period: {tsla_data.index.min()} to {tsla_data.index.max()}")
    print(f"  Rows: {len(tsla_data)}")
    print(f"  Latest Price: ${tsla_data['Close'].iloc[-1]:.2f}")
except Exception as e:
    print(f"✗ Failed to load data: {e}")
    raise
```

```python
# Prepare data with train/test split
series = tsla_data['Close']

# Chronological split: 80% train, 20% test
train_size = int(len(series) * 0.8)
train, test = series[:train_size], series[train_size:]

print(f"\nData Split:")
print(f"  Training: {len(train)} rows ({train.index.min()} to {train.index.max()})")
print(f"  Test: {len(test)} rows ({test.index.min()} to {test.index.max()})")
```

```python
# Fit ARIMA(2,1,2) - best model from Task 2
def fit_arima_model(train_data, order=(2,1,2)):
    """Fit ARIMA model with error handling."""
    try:
        model = ARIMA(train_data, order=order)
        fitted = model.fit()
        logger.info(f"✓ ARIMA{order} fitted successfully")
        return fitted
    except Exception as e:
        logger.error(f"✗ ARIMA model fitting failed: {e}")
        raise

try:
    fitted_model = fit_arima_model(train, order=(2,1,2))
    print(f"\nARIMA(2,1,2) Summary:")
    print(f"  AIC: {fitted_model.aic:.2f}")
    print(f"  BIC: {fitted_model.bic:.2f}")
except Exception as e:
    print(f"✗ Model fitting failed: {e}")
    raise
```

```python
# Generate test predictions (for validation)
test_predictions = fitted_model.forecast(len(test))
test_mae = mean_absolute_error(test, test_predictions)
test_rmse = np.sqrt(mean_squared_error(test, test_predictions))

print(f"\nTest Set Performance:")
print(f"  MAE: ${test_mae:.2f}")
print(f"  RMSE: ${test_rmse:.2f}")
print(f"  MAPE: {np.mean(np.abs((test - test_predictions) / test)) * 100:.2f}%")
```

```python
# Generate 6-month (180 days) and 12-month (360 days) future forecasts
def generate_forecast(fitted_model, steps, alpha=0.05):
    """Generate forecast with confidence intervals."""
    try:
        forecast = fitted_model.get_forecast(steps=steps)
        forecast_mean = forecast.predicted_mean
        forecast_ci = forecast.conf_int(alpha=alpha)
        return forecast_mean, forecast_ci
    except Exception as e:
        logger.error(f"✗ Forecast generation failed: {e}")
        raise

try:
    # 6-month forecast
    forecast_6m_mean, forecast_6m_ci = generate_forecast(fitted_model, 180)
    future_dates_6m = pd.date_range(start=series.index[-1] + pd.Timedelta(days=1), periods=180, freq='B')
    
    # 12-month forecast
    forecast_12m_mean, forecast_12m_ci = generate_forecast(fitted_model, 360)
    future_dates_12m = pd.date_range(start=series.index[-1] + pd.Timedelta(days=1), periods=360, freq='B')
    
    print(f"\n6-Month Forecast:")
    print(f"  Forecasted Price: ${forecast_6m_mean.iloc[-1]:.2f}")
    print(f"  95% CI: [${forecast_6m_ci.iloc[-1, 0]:.2f}, ${forecast_6m_ci.iloc[-1, 1]:.2f}]")
    
    print(f"\n12-Month Forecast:")
    print(f"  Forecasted Price: ${forecast_12m_mean.iloc[-1]:.2f}")
    print(f"  95% CI: [${forecast_12m_ci.iloc[-1, 0]:.2f}, ${forecast_12m_ci.iloc[-1, 1]:.2f}]")
except Exception as e:
    print(f"✗ Forecast generation failed: {e}")
    raise
```

```python
# Plot: Historical, Test Predictions, and Future Forecast (6-month)
fig, ax = plt.subplots(figsize=(16, 8))

# Historical data (last 500 days for clarity)
historical_plot = series.iloc[-500:]
ax.plot(historical_plot.index, historical_plot, 
        label='Historical Data', color='blue', linewidth=2, alpha=0.8)

# Test set actuals
ax.plot(test.index, test, 
        label='Test Actuals', color='green', linewidth=1.5, alpha=0.7)

# Test predictions
ax.plot(test.index, test_predictions, 
        label='Test Predictions', color='orange', linewidth=2, linestyle='--')

# Future forecast
ax.plot(future_dates_6m, forecast_6m_mean, 
        label='6-Month Forecast', color='red', linewidth=2, linestyle=':')

# Confidence intervals for forecast
ax.fill_between(future_dates_6m, 
                 forecast_6m_ci.iloc[:, 0], 
                 forecast_6m_ci.iloc[:, 1], 
                 color='red', alpha=0.15, label='95% Confidence Interval')

# Vertical line at forecast start
ax.axvline(x=series.index[-1], color='black', linestyle='-', alpha=0.5, linewidth=1)
ax.text(series.index[-1], ax.get_ylim()[1]*0.95, 'Forecast Start', 
        ha='center', va='top', fontsize=10, rotation=90)

ax.set_title('TSLA: Historical, Test Predictions, and 6-Month Future Forecast', 
             fontsize=16, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Price ($)')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Plot: Full 12-month forecast with confidence intervals
fig, ax = plt.subplots(figsize=(16, 8))

# Historical data (last 500 days)
ax.plot(historical_plot.index, historical_plot, 
        label='Historical Data', color='blue', linewidth=2, alpha=0.8)

# Future forecast - 12 months
ax.plot(future_dates_12m, forecast_12m_mean, 
        label='12-Month Forecast', color='red', linewidth=2, linestyle=':')

# Confidence intervals
ax.fill_between(future_dates_12m, 
                 forecast_12m_ci.iloc[:, 0], 
                 forecast_12m_ci.iloc[:, 1], 
                 color='red', alpha=0.15, label='95% Confidence Interval')

# Vertical line
ax.axvline(x=series.index[-1], color='black', linestyle='-', alpha=0.5, linewidth=1)

ax.set_title('TSLA: 12-Month Future Forecast with Confidence Intervals', 
             fontsize=16, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Price ($)')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Forecasted return calculation
latest_price = series.iloc[-1]
forecast_6m_price = forecast_6m_mean.iloc[-1]
forecast_12m_price = forecast_12m_mean.iloc[-1]

forecast_6m_return = (forecast_6m_price / latest_price) - 1
forecast_12m_return = (forecast_12m_price / latest_price) - 1

annualized_6m_return = (1 + forecast_6m_return) ** (252/180) - 1

print("="*60)
print("FORECASTED RETURNS")
print("="*60)
print(f"\nLatest Price: ${latest_price:.2f}")
print(f"\n6-Month Forecast:")
print(f"  Forecasted Price: ${forecast_6m_price:.2f}")
print(f"  Total Return: {forecast_6m_return*100:.2f}%")
print(f"  Annualized Return: {annualized_6m_return*100:.2f}%")
print(f"\n12-Month Forecast:")
print(f"  Forecasted Price: ${forecast_12m_price:.2f}")
print(f"  Total Return: {forecast_12m_return*100:.2f}%")
```
